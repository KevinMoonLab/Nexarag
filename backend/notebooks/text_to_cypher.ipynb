```python


def run_nl_query(question: str) -> dict:
    """
    Stateless: NL -> Cypher -> Neo4j rows.
    Returns:
      {
        "cypher": str,
        "rows": list[dict],
      }
    """
    out = cypher_chain.invoke({"query": question})

    # When return_direct=True + return_intermediate_steps=True,
    # you'll typically have intermediate_steps with the query + rows.
    cypher = out["intermediate_steps"][0]["query"]
    rows   = out["intermediate_steps"][0]["context"]

    return {
        "cypher": cypher,
        "rows": rows,
    }

# Usage
result = run_nl_query("List the 10 most recent tracking events for asset 1234.")
print("Cypher:\n", result["cypher"])
print("Rows:\n", result["rows"])
```
